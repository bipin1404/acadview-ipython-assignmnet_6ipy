```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pl
d=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Weather/master/weather.csv")
print(d.head())
```

```
#histogram
sns.distplot(d.MinTemp)
sns.despine()
#sns.distplot(d.Rainfall)
```

```
#barplot
sns.countplot(x='MinTemp',data=d)
sns.despine()
```

```
#boxplot
sns.boxplot(x='MinTemp',y='MaxTemp',data=d)
sns.despine()
```

```
#using title
sns.countplot(x='Temp3pm',data=d)
pl.title("rainfall prediction")
```

```
d['MinTemp'].plot(kind='barh')
```

```
d['Rainfall'].plot(kind='bar')
```

```
d['Rainfall'].plot(kind="pie")
```

```
d["Temp3pm"].plot(kind='hist')
```

```
sns.violinplot(x='Rainfall',data=d)
sns.despine()
```

```
sns.violinplot(y='Temp3pm',data=d)
sns.despine()
```

```
sns.regplot(x='Temp3pm',y='Rainfall',data=d)
sns.despine()
```

```
sns.jointplot(x='Rainfall',y='MaxTemp',data=d)
pl.title("jointplot")
sns.despine()
```
